```python
%matplotlib inline
import os
#downloaded data from https://fr.ftp.opendatasoft.com/infrabel/PunctualityHistory/Raw_punctuality_201903.csv
DATA_PATH = '/Users/kris/data/infrabel'
FILE = 'Raw_punctuality_201903.csv'
FULL_PATH = os.path.join(DATA_PATH, FILE)
db_string = "postgres://kris@localhost/infrabel"
```

```python
import pandas as pd
df = pd.read_csv(FULL_PATH)
df = df[df['DATDEP'] == '04MAR2019']
df
```

```python
df.columns
```

```python
dates = ['DATDEP', 'PLANNED_DATE_ARR', 'PLANNED_DATE_DEP','REAL_DATE_ARR', 'REAL_DATE_DEP']
for date in dates:
    df[date] = pd.to_datetime(df[date]).dt.date
times = ['REAL_TIME_ARR', 'REAL_TIME_DEP', 'PLANNED_TIME_ARR','PLANNED_TIME_DEP']
for time in times:
    df[time] = pd.to_datetime(df[time], format='%H:%M:%S').dt.time
```

```python
df
```

```python
from sqlalchemy import create_engine
engine = create_engine(db_string, echo=False)
df.to_sql('delays', con=engine, if_exists='replace')
```

```python
enriched = """
select *, lag("PTCAR_LG_NM_NL") OVER (PARTITION BY "DATDEP", "TRAIN_NO" ORDER BY index)          as PREV_STOP,
          lag("DELAY_ARR") OVER (PARTITION BY "DATDEP", "TRAIN_NO" ORDER BY index)               as PREV_DELAY_ARR,
          concat("PTCAR_LG_NM_NL", ' -> ',
                 lag("PTCAR_LG_NM_NL") OVER (PARTITION BY "DATDEP", "TRAIN_NO" ORDER BY index))  AS SEGMENT,
          "DELAY_ARR" - lag("DELAY_ARR") OVER (PARTITION BY "DATDEP", "TRAIN_NO" ORDER BY index) AS ADDITIONAL_DELAY,
          date_part('hour', "REAL_TIME_ARR") AS HOUR_REAL_TIME_ARR
from delays
"""
df = pd.read_sql(enriched, con=engine)
df.to_sql('enriched', con=engine, if_exists='replace', index=False)
```

```python
df
```

```python
aggregated = """
select segment, COALESCE(avg(ADDITIONAL_DELAY), 0) AS avg_additional_delay, count(*) as FREQ,
       avg("DELAY_ARR") AS avg_delay
from enriched
where PREV_STOP is not null
group by SEGMENT
order by FREQ desc;
"""
df = pd.read_sql(aggregated, con=engine)
stats = df[(df['freq'] >= 5) & (df['avg_delay'] <= 500) & (df['avg_additional_delay'] >= -150)]
#df = stats[stats['internalLinenumber'] == '3370']
from matplotlib.colors import LogNorm
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")

# Draw a scatter plot while assigning point colors and sizes to different
# variables in the dataset
f, ax = plt.subplots(figsize=(13, 13))
sns.despine(f, left=True, bottom=True)
sns.scatterplot(x="avg_delay", y="avg_additional_delay",
                #hue="tripStopIndex", 
                size="freq",
                sizes=(1, 200), linewidth=0,
                data=stats, ax=ax)
```
